```python
# Line Plot
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [2, 4, 1, 7]

# Create the plot
plt.plot(x, y)
plt.plot(x,y,color='red')
# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

# Display the plot
plt.show()
```

```python
# Pie Chart using numpy
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show()
```

```python
# Pie Chart using numpy  with labels
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
Subjectnames = ["English", "History", "Biology", "Mathematics"]
mycolors = ["black", "blue", "red", "yellow"]
plt.pie(y,labels=Subjectnames,colors=mycolors)    # array, labels, colors
plt.show()
```

```python
# Simpe bar chart
x = ["Pizza", "BANANAS"]
y = [400, 350]
#plt.bar(x, y)
plt.bar(x,y,color='red', width=0.1)
plt.xlabel("Fruits")
plt.ylabel("Quantity")
```

```python
# Barplots
import numpy as np
import matplotlib.pyplot as plt
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30, 'Python':35}
courses = list(data.keys())
values = list(data.values())
# creating the bar plot
plt.bar(courses, values, color ='maroon', width = 0.4)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt
import numpy as np
array = np.array([23, 56, 87, 87, 98,
             12, 76, 98, 34, 87,
             67, 23, 87, 56, 34,
             26, 85, 47, 35, 86,
             76, 45, 86, 34, 40])
# Plotting a basic histogram , explicitly bin ranges are given
plt.hist(array, bins=[20, 40, 60, 80, 100], color='skyblue', edgecolor='black')
# 20 is the initial value, 40 is the outer range, only initial value is included
# (20,40) , (40,60)  (60, 80) (80,100)
# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```
